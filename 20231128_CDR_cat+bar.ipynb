```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv(r'D:\DUIH-AWS\1st_csv\all_f norm by eTIV.csv', index_col=False)
df2 = pd.read_csv(r'D:\git\EDA_chunks\ADNI(with CDR&MMSE Total).T.csv', index_col=False)
```

```python
# Extract 'CDR' column from each DataFrame
cdr_df1 = df1['CDR']
cdr_df2 = df2['CDR']

# Count the occurrences of each class in the 'CDR' column
count_df1 = cdr_df1.value_counts().sort_index()
count_df2 = cdr_df2.value_counts().sort_index()

# Create a grouped bar plot for df1
sns.barplot(x=count_df1.index, y=count_df1.values, color='blue', label='df1')

# Create a grouped bar plot for df2
sns.barplot(x=count_df2.index, y=count_df2.values, color='orange', label='df2')

# Display the legend and labels
plt.legend(title='DataFrames')
plt.xlabel('CDR Class')
plt.ylabel('Count')
plt.title('CDR Class Counts for df1 and df2')

# Show the plot
plt.show()

```

```python

# Add a 'DataFrame' column to identify the source of the data
df1['DataFrame'] = 'DUIH'
df2['DataFrame'] = 'df2'

# Combine the data from df1 and df2 and reset the index
combined_df = pd.concat([df1[['CDR', 'DataFrame']], df2[['CDR', 'DataFrame']]]).reset_index(drop=True)

# Create a catplot for the combined DataFrame
sns.catplot(x='CDR', kind='count', hue='DataFrame', data=combined_df)

# Display the labels
plt.xlabel('CDR Class')
plt.ylabel('Count')
plt.title('Combined CDR Class Counts for df1 and df2')

# Show the plot
plt.show()
```

```python
# Combine the 'CDR' columns from both DataFrames
combined_cdr = pd.concat([cdr_df1, cdr_df2])

# Count the occurrences of each class in the combined 'CDR' column
count_combined = combined_cdr.value_counts().sort_index()

# Create a bar plot for the combined DataFrame
sns.barplot(x=count_combined.index, y=count_combined.values, color='green')
plt.xlabel('CDR Class')
plt.ylabel('Count')
plt.title('Combined CDR Class Counts')

# Show the plot
plt.show()
```
